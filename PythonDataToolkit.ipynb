2. How does broadcasting work in NumPy?

It "stretches" or "replicates" the smaller array across the larger one to make their shapes compatible for element-wise operations. This is a highly efficient process, as the duplication is conceptual rather than actual, saving memory and speeding up computations.

6. What are the differences between NumPy arrays and Python lists?
- The differnece between NumPy arrays and Python lists are:
- 1] Data Type Homogeneity:
The most significant difference is in the data types they can hold. NumPy arrays are homogeneous, meaning all elements within a single NumPy array must be of the same data type (e.g., all integers, all floating-point numbers, all booleans). This strictness allows NumPy to store data very efficiently in contiguous blocks of memory. In contrast, Python lists are heterogeneous, capable of storing elements of different data types within the same list (e.g., [1, 'hello', 3.14, True]).
- 2] Performance and Efficiency:
NumPy arrays offer significantly higher performance and memory efficiency, especially for large datasets and numerical operations. Because NumPy arrays are homogeneous and stored contiguously, operations can be vectorized and executed by highly optimized C and Fortran code, bypassing Python's slower interpretation. This leads to dramatic speedups in mathematical computations. Python lists, on the other hand, are less efficient for numerical tasks. Element-wise operations on lists often require explicit Python loops, which are much slower due to the overhead of type-checking and object lookup for each element.
- 3] Functionality for Numerical Operations:
NumPy is purpose-built for numerical and scientific computing. It provides a vast collection of optimized mathematical functions (universal functions or ufuncs) that operate directly on entire arrays, including linear algebra, Fourier transforms, random number generation, and sophisticated statistical operations. Python lists offer basic list manipulation methods (append, insert, remove, sort, etc.), but they lack the extensive numerical capabilities inherent in NumPy. Performing complex mathematical operations on Python lists typically requires custom functions and loops.
- 4] Memory Storage:
NumPy arrays store their elements in contiguous blocks of memory. This "locality of reference" is a key reason for their speed, as it allows for efficient caching and direct memory access. Conversely, elements in a Python list are stored as references (pointers) to objects that can be located anywhere in memory. This scattered storage contributes to higher memory consumption and slower access times, especially for large datasets.

The variation in color can be by hue (e.g., from blue to red) or by intensity (e.g., from light to dark shades of a single color), allowing for a quick visual understanding of patterns, distributions, and relationships in data.

- 1] Visualize Correlations between Variables:
One of the most common and powerful uses of heatmaps is to display a correlation matrix. Each cell in the heatmap represents the correlation coefficient between two variables, with the color indicating the strength and direction (positive or negative) of that correlation. This allows you to quickly identify highly correlated or inversely correlated pairs of variables in a large dataset, which is crucial for feature selection in machine learning or understanding multicollinearity.


- 2] Display Data in a Matrix or Grid Format:
If your data naturally fits into a two-dimensional grid, a heatmap is an excellent choice. This includes:

- 3] Tabular Data:
When you want to visualize a numerical table, a heatmap makes it easy to spot "hot spots" (high values) and "cold spots" (low values) without having to read every number.

- 4] Time Series Data with Periodicity: Heatmaps are great for showing patterns over time, such as daily sales across different weeks, website traffic by hour of the day and day of the week, or seasonal trends. The axes would represent time periods (e.g., days of the week vs. hours of the day).

8. What does the term “vectorized operation” mean in NumPy?
- In NumPy, "vectorized operation" refers to the process of applying mathematical operations or functions to entire arrays (or "vectors" in a broader sense of a sequence of numbers) at once, rather than iterating over individual elements using explicit Python loops. This concept is a cornerstone of NumPy's efficiency and performance.

17. Explain the difference between apply() and map() in Pandas.
- In Pandas, both apply() and map() are powerful functions used for applying custom operations to data within Series and DataFrames. However, they differ significantly in their primary use cases, flexibility, and how they interact with data structures.

- The map() function is defined only on a Pandas Series. Its primary purpose is to map values from one Series to another set of values based on a mapping dictionary, a Series, or a function. It's essentially designed for element-wise substitution or transformation.
Key characteristics of map():
- 1] Operates on a Series only: You call map() directly on a Pandas Series.

- 2] Element-wise operation: It applies a function or mapping operation to each individual element of the Series.

- The apply() function is much more versatile and can be called on both Series and DataFrames. Its role is to apply a function along an axis of a DataFrame or Series. This function can operate element-wise, row-wise, or column-wise.
Key characteristics of apply():

- 1] Operates on Series or DataFrame:
On a Series, it behaves similarly to map() for element-wise operations, but with more flexibility (e.g., returning multiple values).

- 2] On a DataFrame, it applies a function to each row (axis=1) or each column (axis=0).

18. What are some advanced features of NumPy?
- 1] Broadcasting: This is a fundamental and incredibly powerful feature that allows NumPy to perform arithmetic operations on arrays of different shapes and sizes without explicitly making copies of data. It conceptually "stretches" the smaller array across the larger one to make their shapes compatible. This mechanism significantly reduces memory consumption and speeds up computations, as the duplication is virtual rather than actual.
- 2] Memory Management (Views vs. Copies) and Strides:
NumPy is highly optimized to avoid unnecessary data copying. Many operations (like slicing or reshaping) return "views" into the original array's data, meaning they share the same underlying memory. This saves memory and improves performance. Understanding strides which define the number of bytes to jump in memory to move to the next element along each dimension is key to comprehending how NumPy efficiently handles multidimensional arrays internally and how operations like reshape can be performed without copying data.
- 3] Structured Arrays (Record Arrays):
NumPy allows the creation of "structured arrays" (also known as record arrays or heterogeneous arrays) where different elements in the same array can have different data types and can be accessed by field names, similar to records in a database or structs in C. This is particularly useful for storing tabular data where each row is a record with named fields of varying types. While Pandas DataFrames are often preferred for this, structured arrays can offer performance benefits in specific scenarios or for interfacing with C/Fortran code.

21. Why is NumPy's array slicing faster than Python's list slicing?

- NumPy array slicing is significantly faster than Python's built-in list slicing due to fundamental differences in how they are implemented and how they handle data in memory. Here's a breakdown of the key reasons:

- 1] Contiguous Memory Allocation (NumPy Arrays):

- NumPy: ndarray elements are stored in a contiguous block of memory. This means all elements are packed tightly next to each other. When you slice a NumPy array, it often returns a "view" of the original array, rather than creating a new copy of the data. This "view" is essentially a new ndarray object that simply points to a different starting address in the original memory block and updates its shape and strides (rules for how to jump through memory to find elements). This avoids the overhead of memory allocation and data copying.

- Python Lists: Python lists are implemented as arrays of pointers. Each element in a list is an independent Python object that can be of any data type (heterogeneous). The list itself stores only references (pointers) to these objects, and these objects can be scattered throughout memory. When you slice a Python list, it always creates a new list object and copies the references to the new memory location. Even if the elements themselves aren't copied, the references are, which incurs a significant overhead for large lists.

- 2] Homogeneous Data Type (NumPy Arrays):

- NumPy: All elements in a NumPy array must be of the same data type. This homogeneity allows NumPy to use highly optimized, C-level routines for operations. When slicing, NumPy knows the exact size of each element, making memory access direct and efficient.

- Python Lists: Python lists can store heterogeneous data types. This flexibility comes at a cost. Python needs to store type information and reference counts for each individual object in the list. This overhead means more memory consumption per element and slower operations because the interpreter must constantly check the type of each element during operations.


- 3] Underlying C/Fortran Implementation (NumPy):

- NumPy: The core operations of NumPy, including array slicing and other array manipulations, are implemented in highly optimized C and Fortran code. When you perform a slice, it's not executed by Python's slower interpreter but by pre-compiled, fast machine code.

- Python Lists: List slicing is handled by Python's C-implemented core, but it still operates on Python objects and involves the overhead of Python's object model.

```python
# 3. Create a scatter plot using Matplotlib

import pandas as pd
import matplotlib.pyplot as plt # Import Matplotlib

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'John'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85, 92, 78, 95, 88],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n" + "="*30 + "\n")

# Specify the column for which you want to calculate the mean
column_to_mean = 'Score'

# Calculate the mean of the specified column
mean_value = df[column_to_mean].mean()

# Print the result
print(f"The mean of the '{column_to_mean}' column is: {mean_value}")

# Example with another column
column_to_mean_age = 'Age'
mean_age = df[column_to_mean_age].mean()
print(f"The mean of the '{column_to_mean_age}' column is: {mean_age}")

print("\n" + "="*30 + "\n")

# Create a scatter plot using Matplotlib
print("Creating a scatter plot of Age vs. Score: \n")
plt.figure(figsize=(5, 2))  # This creates a new figure and sets its size.
plt.scatter(df['Age'], df['Score'], color='violet', alpha=1) # Create the scatter plot
plt.title('Age vs. Score') # Add a title to the plot
plt.xlabel('Age') # Label the x-axis
plt.ylabel('Score') # Label the y-axis
plt.grid(True, linestyle='-', alpha=1) # Add a grid for better readability
plt.show() # Display the plot

```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

import pandas as pd # Import Pandas
import matplotlib.pyplot as plt # Import Matplotlib
import seaborn as sns # Import Seaborn for heatmap

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'John'],
    'Age': [24, 27, 22, 32, 29],
    'Score': [85, 92, 78, 95, 88],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney']
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n" + "="*30 + "\n")

# Specify the column for which you want to calculate the mean
column_to_mean = 'Score'

# Calculate the mean of the specified column
mean_value = df[column_to_mean].mean()

# Print the result
print(f"The mean of the '{column_to_mean}' column is: {mean_value}")

# Example with another column
column_to_mean_age = 'Age'
mean_age = df[column_to_mean_age].mean()
print(f"The mean of the '{column_to_mean_age}' column is: {mean_age}")

print("\n" + "="*30 + "\n")

# Create a scatter plot using Matplotlib
print("Creating a scatter plot of Age vs. Score: \n")
plt.figure(figsize=(5, 2)) # This creates a new figure and sets its size.
plt.scatter(df['Age'], df['Score'], color='violet', alpha=1) # Create the scatter plot
plt.title('Age vs. Score') # Add a title to the plot
plt.xlabel('Age') # Label the x-axis
plt.ylabel('Score') # Label the y-axis
plt.grid(True, linestyle='--', alpha=0.6) # Add a grid for better readability
plt.show() # Display the plot

print("\n" + "="*30 + "\n")

# Calculate the correlation matrix and visualize it with a heatmap using Seaborn
print("Calculating and visualizing the correlation matrix")

# Select only numerical columns for correlation calculation
numerical_df = df.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)
print()

# Create a heatmap
plt.figure(figsize=(5, 2)) # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of Numerical Columns') # Add a title
plt.show() # Display the heatmap
```

```python
# 5. Generate a bar plot using Plotly.

import pandas as pd
import plotly.express as px

data = {
    'Category': ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Pears'],
    'Sales': [150, 200, 120, 180, 90]
}
df = pd.DataFrame(data)

# Create a bar plot
fig = px.bar(
    df,
    x='Category', # x: column to be used for the x-axis (categories)
    y='Sales',  # y: column to be used for the y-axis (values)
    title='Fruit Sales by Category',  # title: title of the plot
    labels={'Category': 'Fruit Type', 'Sales': 'Units Sold'}, # labels: dictionary to customize axis labels
    color='Category' # Color bars differently based on category
)

fig.update_layout(
    xaxis_title_font_size=20,
    yaxis_title_font_size=20,
    title_font_size=20,
    plot_bgcolor='white', # Transparent background
    paper_bgcolor='black', # Transparent paper background
    width=700,  # Set the width of the figure
    height=500 # Set the height of the figure
)

fig.show()
```

```python
 # 6. Create a DataFrame and add a new column based on an existing column.

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Sample data for the bar plot
data = {
    'Category': ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Pears'],
    'Sales': [150, 200, 120, 180, 90]
}
df = pd.DataFrame(data)

# Add a new column 'Profit' based on the 'Sales' column
# For example, let's say profit is 20% of sales
df['Profit'] = df['Sales'] * 0.20

# Create a bar plot for Sales
fig = go.Figure(data=[
    go.Bar(name='Sales', x=df['Category'], y=df['Sales']),
    go.Bar(name='Profit', x=df['Category'], y=df['Profit'])
])

# Update layout for grouped bar chart
fig.update_layout(
    barmode='group', # Set bar mode to 'group'
    title='Fruit Sales and Profit by Category',
    xaxis_title='Fruit Type',
    yaxis_title='Amount',
    xaxis_title_font_size=14,
    yaxis_title_font_size=14,
    title_font_size=20,
    plot_bgcolor='white',
    paper_bgcolor='black',
    width=700,
    height=500
)

fig.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.

import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [24, 30, 22, 35, 28],
    'Score': [85, 92, 78, 95, 88]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
print("\n" + "=" *50 + "\n")
# Define a threshold
score_threshold = 90

# Filter rows where the 'Score' column is greater than the threshold
# This creates a boolean Series, which is then used to select rows
filtered_df = df[df['Score'] > score_threshold]

print(f"DataFrame filtered where 'Score' is greater than {score_threshold}:")
print(filtered_df)

print("\n" + "=" *50 + "\n")

# Filtering by 'Age'
age_threshold = 25
filtered_by_age_df = df[df['Age'] > age_threshold]

print(f"DataFrame filtered where 'Age' is greater than {age_threshold}:")
print(filtered_by_age_df)

```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a sample DataFrame or Series with numerical data
data = {
    'Value': np.random.normal(loc=50, scale=10, size=1000)
    # Generates 1000 random numbers from a normal distribution with a mean (loc) of 50 and standard deviation (scale) of 10.
}
df = pd.DataFrame(data)

print("Sample Data (first 5 rows):")
print(df.head())
print()

# 2. Create a histogram using Seaborn's histplot
#    - 'data': The DataFrame containing the data.
#    - 'x': The column to plot on the x-axis (the distribution).
#    - 'kde': Set to True to overlay a Kernel Density Estimate (KDE) curve.
#    - 'bins': Number of bins or a list of bin edges (optional, Seaborn auto-determines if not specified).
#    - 'color': Color of the bars.
#    - 'edgecolor': Color of the bar edges.
plt.figure(figsize=(10, 6)) # Set the figure size for better visualization
sns.histplot(data=df, x='Value', kde=True, bins=30, color='pink', edgecolor='red')

# Add labels and title for clarity
plt.title('Distribution of Sample Values (Histogram with KDE)', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.8)

# Display the plot
plt.show()

```

```python
# 11. Perform matrix multiplication using NumPy.

import numpy as np

# Define two simple 2D NumPy arrays (matrices)

# Matrix A (2x2 matrix)
matrix_a = np.array([
    [1, 2],
    [3, 4]
])

# Matrix B (2x2 matrix)
matrix_b = np.array([
    [5, 6],
    [7, 8]
])

print("Matrix A:")
print(matrix_a)
print("\nMatrix B:")
print(matrix_b)

# Perform matrix multiplication using the @ operator
result_matrix = matrix_a @ matrix_b

print("\nResult of matrix multiplication:")
print(result_matrix)
```

```python
# 12. Use Pandas to load a CSV file and display its first 5 rows.

import pandas as pd

# Define the filename
csv_filename = "simple_data.csv"

# Create some simple data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'City': ['New York', 'London', 'Paris', 'Tokyo', 'Sydney']
}

# Convert the data to a Pandas DataFrame
sample_df = pd.DataFrame(data)

# Create the CSV file from the DataFrame
# index=False prevents writing the DataFrame's row numbers into the CSV
print(f"Creating '{csv_filename}'")
sample_df.to_csv(csv_filename, index=False)
print("CSV file created successfully!")

# Load the contents of the CSV file back into a new DataFrame
print(f"\nLoading '{csv_filename}'")
loaded_df = pd.read_csv(csv_filename)
print("CSV file loaded successfully!")

# Display the first 5 rows of the loaded DataFrame
print("\nFirst 5 rows of the loaded DataFrame:")
print(loaded_df.head(5))
```

```python
# 13. Create a 3D scatter plot using Plotly.

import plotly.graph_objects as go
import numpy as np

# Generate some sample 3D data
np.random.seed(42)
num_points = 100
x = np.random.rand(num_points) * 10
y = np.random.rand(num_points) * 10
z = np.random.rand(num_points) * 10

#  Add a fourth dimension for color based on 'z' value
colors = z

# Create the 3D scatter plot trace
# mode='markers' ensures it's a scatter plot
scatter_trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=5,
        color=colors,  # Set color based on the 'colors' array
        colorscale='Viridis', # Choose a colorscale
        opacity=0.8,
        colorbar=dict(title='Z Value') # Add a color bar legend
    )
)

# Create the figure object and add the trace
fig = go.Figure(data=[scatter_trace])

# Update the layout for titles and axis labels
fig.update_layout(
    title='Interactive 3D Scatter Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

# Display the plot
fig.show()
```
